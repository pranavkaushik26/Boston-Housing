```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('../input/bostonhoustingmlnd/housing.csv')
```

```python
data.info()
```

The target variable which we need to predict is 'MEDV'

```python
sns.heatmap(data.corr())
```

```python
X = data.drop('MEDV',axis=1)
Y = data['MEDV']
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(X,Y)
```

# **1. Linear Regression**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
np.mean(cross_val_score(lr, X, Y, cv=5))
```

# 2. **Ridge Regression**

```python
from sklearn.linear_model import RidgeCV
ridge = RidgeCV(cv=5).fit(X, Y)
ridge.score(X, Y)
```

# **2. Lasso Regression**

```python
from sklearn.linear_model import LassoCV
lasso = LassoCV(cv=5).fit(X, Y)
lasso.score(X, Y)
```

# **4. Polynomial Regression**

```python
from sklearn.preprocessing import PolynomialFeatures

def create_polynomial_regression_model(degree):
    poly_features = PolynomialFeatures(degree=degree)
    X_poly = poly_features.fit_transform(X)
    poly = LinearRegression()
    return np.mean(cross_val_score(poly, X_poly, Y, cv=5))

poly_cv = []
for i in range(1,4):
    poly_cv.append(create_polynomial_regression_model(i))

plt.scatter(range(1,4),poly_cv)
```

```python
create_polynomial_regression_model(2)
```

# **5. KNN Regression**

```python
from sklearn.neighbors import KNeighborsRegressor

cv_score=[]
for i in range(1,10):
    knn = KNeighborsRegressor(n_neighbors= i)
    cv_score.append(np.mean(cross_val_score(knn,X,Y,cv=5)))

x = range(1,10)
plt.scatter(x,cv_score)
```
